```python
print(
    ggplot(mtcars, aes('mpg', 'qsec')) + \
    geom_point(colour = 'steelblue') + \
    scale_x_continuous(breaks = [10, 20, 30], labels = ['Bad', 'Pk', 'Good']));
```

```python
print(
    ggplot(meat, aes('date', 'beef')) + \
    geom_line(color = 'black') + \
    scale_x_date(breaks = date_breaks('7 years'), labels = '%b %Y') + \
    scale_y_continuous(labels = 'comma'));
```

```python
yearLocator = YearLocator(7)
dateFormatter = DateFormatter('%b %Y')

x = meat.date
y = meat.beef

fig, ax = plt.subplots()

ax.plot(x, y, 'black')
ax.xaxis.set_major_locator(yearLocator)
ax.xaxis.set_major_formatter(dateFormatter)

fig.autofmt_xdate()

plt.show()
```
