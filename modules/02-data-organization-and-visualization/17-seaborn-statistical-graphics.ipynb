### Building Statistical Graphs

**distplot()**

```python
sb.distplot(iris.sepal_length, rug = True, fit = stats.gausshyper);
```

**jointplot()**

```python
sb.jointplot(x = 'sepal_length', y = 'petal_length', data = iris);
```

```python
with sb.axes_style('white'):
    sb.jointplot(x = 'sepal_length', y = 'petal_length', data = iris, kind = 'hex', color = 'k');
```

```python
sb.jointplot(x = 'sepal_length', y = 'petal_length', data = iris, kind = 'kde');
```

```python
graphic = sb.jointplot(x = 'sepal_length', y = 'petal_length', data = iris, kind = 'kde', color = 'm');
graphic.plot_joint(plt.scatter, c = 'w', s = 30, linewidth = 1, marker = '+');
graphic.ax_joint.collections[0].set_alpha(0);
```

**pairplot()**

```python
sb.pairplot(iris);
```

### Linear Relationship Visualization

**Loading Datasets**

```python
tips = sb.load_dataset('tips');
anscombe = sb.load_dataset('anscombe');
```

**jointplot()**

```python
sb.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'reg');
```

**lmplot()**

```python
sb.lmplot(x = 'total_bill', y = 'tip', data = tips);
```

```python
sb.lmplot(x = 'size', y = 'tip', data = tips, x_jitter = .05);
```

```python
sb.lmplot(x = 'size', y = 'tip', data = tips, x_estimator = np.mean);
```

```python
sb.lmplot(x = 'x', y = 'y', data = anscombe.query("dataset == 'II'"), ci = None, scatter_kws = {'s': 80});
```

```python
sb.lmplot(x = 'x', y = 'y', data = anscombe.query("dataset == 'II'"), order = 2, ci = None, scatter_kws = {'s': 80});
```

```python
sb.lmplot(x = 'x', y = 'y', data = anscombe.query("dataset == 'III'"), ci = None, scatter_kws = {'s': 80});
```

```python
sb.lmplot(x = 'total_bill', y = 'tip', data = tips, lowess = True);
```

```python
sb.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = tips);
```

```python
sb.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = tips, markers = ['o', 'x'], palette = 'Set1');
```

```python
sb.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'time', data = tips);
```

```python
sb.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'time', row = 'sex', data = tips);
```

```python
sb.lmplot(x = 'total_bill', y = 'tip', col = 'day', data = tips, col_wrap = 2, size = 3);
```

```python
sb.lmplot(x = 'total_bill', y = 'tip', col = 'day', data = tips, aspect = .5);
```
