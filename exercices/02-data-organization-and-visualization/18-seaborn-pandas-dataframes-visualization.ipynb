```python
import matplotlib.pyplot as plt
import pandas as pd
import random
import seaborn as sb

%matplotlib inline
```

```python
dataframe = pd.DataFrame()
```

```python
dataframe['x'] = random.sample(range(1, 100), 25)
dataframe['y'] = random.sample(range(1, 100), 25)
```

```python
sb.lmplot('x', 'y', data = dataframe, fit_reg = False);
```

```python
sb.kdeplot(dataframe.y);
```

```python
sb.distplot(dataframe.x);
```

```python
plt.hist(dataframe.x, alpha = .3)
sb.rugplot(dataframe.x);
```

```python
sb.heatmap([dataframe.y, dataframe.x], annot = True, fmt = 'd');
```

```python
sb.clustermap(dataframe);
```
